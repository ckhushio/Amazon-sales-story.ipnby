```python
df=data.drop(['Order ID','SKU','ship-city','ship-postal-code','Unnamed: 22','ship-country','ASIN','Courier Status','currency','fulfilled-by'],axis=1)
df.sample(10)
```

```python
dfc=df.drop(['Style'],axis=1)
```

```python
dfc
```

```python
fill_values={'Amount':0}
dfc.fillna(fill_values,inplace=True)
```

```python
dfc.Status.value_counts().plot(kind='bar')
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Counts of status of order')
plt.show()
```

```python
dfc.Size.unique()
```

```python
sizes=dfc.Size.value_counts().plot(kind='bar')
plt.xlabel('Size')
plt.ylabel('Count')
plt.title('Counts of Each Size')
plt.show()
```

```python
# checking details about fulfilment status

dfc.Fulfilment.unique()
dfc.Fulfilment.value_counts().plot(kind='bar')
plt.xlabel('Fulfilment')
plt.ylabel('Count')
plt.title('Counts of fulfilment status')
plt.show()
```

```python
dfc.B2B.value_counts().plot(kind='bar')
dfc.B2B.value_counts()
plt.xlabel('B2B')
plt.ylabel('Count')
plt.title('Counts of B2B')
plt.show()
```

Business to business sales are way more than less by amazon which is 871 as expected and other way around which is Business to customer B2C is 128104

```python
dfc.Category.unique()
```

```python
dfc['Amount']=dfc['Amount'].astype(int)
```

```python
# Calculate total sales amount for each state
state_sales = dfc.groupby('ship-state')['Amount'].sum().reset_index()

# Sort states based on sales amount in descending order
state_sales = state_sales.sort_values(by='Amount', ascending=False)

# Select top 10 states
top_10_states = state_sales.head(10)

# Plot barplot for top 10 states
plt.figure(figsize=(15,5))
sns.barplot(x='ship-state', y='Amount', data=top_10_states)
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount for Top 10 States')
plt.show()
```

```python
# Calculate total sales amount for each state
state_sales = dfc.groupby('Category')['Amount'].sum().reset_index()

# Sort states based on sales amount in descending order
state_sales = state_sales.sort_values(by='Amount', ascending=False)

# Select top 10 states
top_10_states = state_sales.head(10)

# Plot barplot for top 10 states
plt.figure(figsize=(15,5))
sns.barplot(x='Category', y='Amount', data=top_10_states)
plt.xlabel('Category')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount for Top 10 Category')
plt.show()
```

Comparing state wise sales where Set has the highest grossing followed by Kurta and then western dreses
